# Clustering

## Number of clusters and K-Means
First we choose the best number of clusters using DB index and Silhouette score

```python
from sklearn.metrics import davies_bouldin_score
from sklearn.metrics import silhouette_score
```

```python
def eval_clustering(X,cluster_func,score_func,**kwargs):

    scores = []

    for k in tqdm(kwargs["k_range"]):
        cluster = cluster_func(k,**kwargs["c_args"])
        labels = cluster.fit_predict(X)
        score = score_func(X,labels,**kwargs["s_args"]) if "s_args" in kwargs else score_func(X,labels)
        scores.append((k, score))

    return scores
```

```python
def plot_scores(scores,k_range):
    ax_x, ax_y = zip(*scores)

    fig = plt.figure(figsize=(10,5))
    ax = plt.axes()
    plt.xticks(k_range)
    plt.grid()
    plt.plot(ax_x, ax_y)
```

As silthuette score takes a lot of time, we'll set the sample size to 10000

### Silhouette score

```python
from tqdm.notebook import tqdm
from sklearn.cluster import KMeans

k_range = range(2,20)

scores = eval_clustering(X_umap,
                         KMeans,
                         silhouette_score,
                         k_range = k_range,
                         c_args={
                             "init":"k-means++",
                             "n_init":1
                         },
                         s_args={
                             "metric":"euclidean",
                             "sample_size":10000
                         })
scores
```

```python
plot_scores(scores,k_range)
```

```python
scores = eval_clustering(X_umap,
                         KMeans,
                         davies_bouldin_score,
                         k_range = k_range,
                         c_args={
                             "init":"k-means++",
                             "n_init":1
                         })
scores
```

```python
plot_scores(scores,k_range)
```

### DB-Scan

First determine distances for min neighbors of 100

```python
from sklearn.preprocessing import MinMaxScaler

data_scaled = MinMaxScaler().fit_transform(data)
```

```python
from sklearn.neighbors import NearestNeighbors

neighbors = NearestNeighbors(n_neighbors=10)
neighbors_fit = neighbors.fit(data_scaled)
distances, indices = neighbors_fit.kneighbors(data_scaled)
```

```python
distances = np.sort(distances, axis=0)
distances = distances[:,1]
plt.plot(distances)
```

Best epsilon is somewhere from 4.5 to 5.0

```python
from sklearn.cluster import DBSCAN

interesting_eps = np.linspace(0.5, 0.6, 8)
score_eps_labels = {}

for eps in tqdm(interesting_eps):
    dbscan = DBSCAN(eps=eps,n_jobs=-1)
    labels = dbscan.fit_predict(data_scaled)

    score = silhouette_score(data_scaled, labels, metric='euclidean', sample_size=300)
    score_eps_labels[score] = (eps, labels)

best_score = max(score_eps_labels)
eps, labels = score_eps_labels[best_score]
print(f"Best score: {best_score}\nParameters: epsilon = {eps}")
sns.scatterplot(X_umap, x=X_umap[:,0], y=X_umap[:,1],hue=labels, s=25)
```

```python
interesting_pts = range(100, 200, 20)
score_pts_labels = {}

for pts in tqdm(interesting_pts):
    dbscan = DBSCAN(eps=eps,min_samples=pts,n_jobs=-1)
    labels = dbscan.fit_predict(data_scaled)

    score = silhouette_score(data_scaled, labels, metric='euclidean', sample_size=500)
    score_pts_labels[score] = (eps, pts, labels)

best_score = max(score_pts_labels)
eps, pts, labels = score_pts_labels[best_score]
print(f"Best score: {best_score}\nParameters: epsilon = {eps}; minPts = {pts}")
sns.scatterplot(X_umap, x=X_umap[:,0], y=X_umap[:,1],hue=labels, s=25)
```
